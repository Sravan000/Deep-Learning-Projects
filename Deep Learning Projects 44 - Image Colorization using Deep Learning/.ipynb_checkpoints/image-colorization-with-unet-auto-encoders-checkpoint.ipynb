```python
''' function to display images '''
def disp(img):
    plt.figure()
    plt.set_cmap('gray')
    plt.imshow(img)
    plt.show()


def combine_lab(l, x, y):
    s = (l.shape[0], l.shape[1], 3)
    z = np.zeros(s)
    z[:, :, 0] = l
    z[:, :, 1] = x
    z[:, :, 2] = y
    return zeros


def combine_ab(x, y):
    s = (a.shape[0], b.shape[1], 2)
    z = np.zeros(s)
    z[:, :, 0] = x
    z[:, :, 1] = y
    return z


def combine_lab(l, xy):
    s = (l.shape[0], l.shape[1], 3)
    z = np.zeros(s)
    z[:, :, 0] = l
    z[:, :, 1] = xy[:, :, 0]
    z[:, :, 2] = xy[:, :, 1]
    return z


def make_channels(gray):
    shape = (gray.shape[0], gray.shape[1], 3)
    zeros = np.zeros(shape)
    zeros[:, :, 0] = gray
    zeros[:, :, 1] = gray
    zeros[:, :, 2] = gray
    return zeros


def get_from_gray(img_path):
    img = io.imread(img_path)
    img = skimage.transform.resize(img,(64,64))
    g = color.rgb2gray(img)
    g = make3channels(f)
    lg = color.rgb2lab(f, illuminant='D50')[:, :, 0]
    return lg


def get_file(f):
    img = io.imread(f)
    xy = np.zeros((64, 64, 2))
    xy[:, :, 0] = img[:, :, 1]
    xy[:, :, 1] = img[:, :, 2]
    return xy


''' function to normalize images '''
def normal_image(p):
    l, xy = load_img_for_training(p)
    l, xy = (l-127.5)/127.5, (xy-127.5)/127.5
    return l, xy


def rgb_img(l, ab):
    s = (l.shape[0],l.shape[1],3)
    img = np.zeros(s)
    img[:,:,0] = l[:,:,0]
    img[:,:,1:]= ab
    img = img.astype('uint8')
    img = cv2.cvtColor(img, cv2.COLOR_LAB2RGB)
    return img

''' function to load training images '''
def load_img_training(img_path):
    img = io.imread(img_path)
    img = skimage.transform.resize(img,(64,64))
    lab = color.rgb2lab(img, illuminant='D50')
    l, a, b = lab[:, :, 0], lab[:, :, 1], lab[:, :, 2]
    ab = combine_ab(a, b)
    lg = get_from_gray(img_path)
    return lg, ab


def save_ab_file(img, fp):
    s = (img.shape[0], img.shape[1], 3)
    new_img = np.zeros(s)
    new_img[:, :, 1] = image[:, :, 0]
    new_img[:, :, 2] = image[:, :, 1]
    save_file(new_img, fp)


''' function to save images '''
def save_file(img, fp):
    io.imsave(fp, img)

''' function to load images '''
def load_image(p):
    img = io.imread(p)
    s = (img.shape[0], img.shape[1], 2)
    xy = np.zeros(s)
    xy[:, :, 0] = img[:, :, 1]
    xy[:, :, 1] = img[:, :, 2]
    return xy
```

```python
def normalize(image):
    ''' scaling an image '''
    img = image/255
    return img

def unnormalize(image):
    ''' unscaling an image '''
    img = (image*255)
    return img.astype('uint8')
```

```python
x = x/256.0
```

```python
IMAGE_SIZE = (224, 224)

''' InceptionResNetV2 Model  '''
model1 = keras.applications.InceptionResNetV2(weights='imagenet', include_top=False ,input_shape=[*IMAGE_SIZE, 3])

''' freezing layers '''
for l in model1.layers:
    l.trainable = False

''' lets see how model looks like '''
model1.summary()

''' encoder '''
encoder_model = model1
```

```python
model1.output_shape[1:4]
```

```python
''' decoder '''
inp = Input(model1.output_shape[1:4])
output = ReLU()(inp)

l_dim = [512, 512, 256, 128, 64, 32]

for i in l_dim:
    ''' Conv2DTranspose Layer '''
    output = Conv2DTranspose(i, (2, 2), strides=(2, 2), padding='same', activation='relu')(output)
    output = Conv2D(i, 2, padding='same', activation='relu')(output)

''' Convolutional Layer '''
output = Conv2D(3,2,padding='same',activation='sigmoid')(output)
output = keras.layers.Lambda(lambda img: tf.image.resize_images(img, (224, 224), method=tf.image.ResizeMethod.BICUBIC, 
                                                                align_corners=True, 
                                                                preserve_aspect_ratio = True))(output)

decoder_model = Model(inputs=[inp], outputs=[output])

''' compile the model '''
decoder_model.compile(optimizer='adam', loss='mse', metrics=['acc'])

''' lets see how decoder model looks like '''
decoder_model.summary()
```

```python
o = decoder_model(encoder_model.output)

''' freezing layers '''
encoder_model.trainable=False

''' final Model '''
model = Model(encoder_model.input, o)

''' compile the model '''
model.compile(optimizer='adam', loss='mae', metrics=['acc'])
''' lets see how final model looks like '''
model.summary()
```

```python
''' training '''
samples = x.shape[0]
epochs = 10
history = model.fit(x,x,validation_split=0.1,epochs=epochs,batch_size=64)

''' saving the model '''
model.save('model.h5')
```
