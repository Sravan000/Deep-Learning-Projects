```python
import pandas as pd 
import warnings 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline

warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
```

```python
''' reading data '''
df_master = pd.read_csv('Covid 19 Dataset/master_results_table.csv')
df_mers = pd.read_csv("Covid 19 Dataset/mers_results.csv")
df_sars = pd.read_csv("Covid 19 Dataset/mers_results.csv")
```

```python
''' count plot of unnamed: 0 '''
plt.figure(figsize=(10, 5))
sns.countplot(df_sars['Unnamed: 0'])
```

```python
''' count plot of connectivity '''
plt.figure(figsize=(10, 5))
sns.countplot(df_sars["Connectivity"]);
```

```python
''' count plot of P_value '''
plt.figure(figsize=(10, 5))
sns.countplot(df_sars['P_value']);
```

```python
''' scatter plot b/w connectivtiy an p_value'''
plt.figure(figsize=(10, 5))
sns.scatterplot(df_sars['Connectivity'], df_sars['P_value']);
plt.title('scatter plot of df_sars b/w connectivity and p_value', fontsize=20)
plt.xlabel('Connectivity', fontsize=15);
plt.ylabel('P_value', fontsize=15);
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df_mers['Connectivity'], df_mers['P_value'])
plt.title('scatter plot of df_mers b/w connectivity and p_value', fontsize=20)
plt.xlabel('Connectivity', fontsize=15);
plt.ylabel('P_value', fontsize=15);
```
